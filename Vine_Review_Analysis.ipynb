```python
# Initial imports
import pandas as pd
```

```python
# Loading data
file_path = "Resources/vine_table.csv"
vine_table = pd.read_csv(file_path)
vine_table.head()
```

```python
vine_table_filt = vine_table[vine_table["total_votes"] >= 20]
```

```python
vine_table_filt
```

```python
# Filter the new DataFrame and create a new DataFrame to retrieve all the 
# rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

vine_table_filt2 = vine_table[(vine_table["helpful_votes"]/vine_table["total_votes"]) >= 0.50]
```

```python
vine_table_filt2
```

```python
# Filter the DataFrame create a new DataFrame or table that 
# retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.

paid_vine_table = vine_table_filt2[vine_table_filt2["vine"] == "Y"]

paid_vine_table
```

```python
# Retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.

unpaid_vine_table = vine_table_filt2[vine_table_filt2["vine"] == "N"]

unpaid_vine_table
```

```python
# Determine the total number of reviews for the two types of review (paid vs unpaid).
paid_count = len(paid_vine_table)
print(paid_count)

unpaid_count = len(unpaid_vine_table)
print(unpaid_count)
```

```python
# Determine the number of 5-star reviews for the two types of review (paid vs unpaid).
paid_fivestar_reviews = paid_vine_table[paid_vine_table["star_rating"] == 5]["star_rating"].count()
print(paid_fivestar_reviews)
unpaid_fivestar_reviews = unpaid_vine_table[unpaid_vine_table["star_rating"] == 5]["star_rating"].count()
print(unpaid_fivestar_reviews)
```

```python
# Determine the percentage of 5-star reviews for the two types of review (paid vs unpaid).
percentage_paid_review = (paid_fivestar_reviews / paid_count) * 100
percentage_unpaid_review = (unpaid_fivestar_reviews / unpaid_count) * 100

print(percentage_paid_review)
print(percentage_unpaid_review)
```
